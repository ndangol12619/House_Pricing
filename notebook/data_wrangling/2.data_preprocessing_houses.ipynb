```python
file_path = pathlib.Path('D:\Capstone Project\dataset\House_Price_dataset')
df = pd.read_csv(f"{file_path}/houses.csv")
df.head()
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
# check for missing values
df.isnull().sum()
```

```python
# Columns to drop -> property_name, link, property_id
df.drop(columns=['link','property_id'], inplace=True)
```

```python
df.head()
```

```python
# rename columns
df.rename(columns={'rate':'price_per_sqft'},inplace=True)
df.head()
```

```python
# society
df['society'].value_counts()
```

```python
df['society'].value_counts().shape
```

```python
## Cleaning Textual Data For Society Column
import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
print(df['society'].value_counts().shape)
df.head()
```

```python
df['society'].value_counts().shape
```

```python
## Replacting nan Value to independent on society column
df['society'] = df['society'].str.replace('nan','independent')
```

```python
# price
df['price'].value_counts()
```

```python
## Filter Out Price on Request from the price column
df = df[df['price'] != 'Price on Request']
df.head()
```

```python
## Assigning Standard Metric for Price Column
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
        
df['price'] = df['price'].str.split(' ').apply(treat_price)
df.head()
```

```python
# price_per_sqft
df['price_per_sqft'].value_counts()
```

```python
## Further Text value Cleaning price_per_sqft column 
df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype('float')
df.head()
```

```python
# bedrooms
df['bedRoom'].value_counts()
```

```python
df[df['bedRoom'].isnull()].head(2)
```

```python
## Filter Out Null Value
df = df[~df['bedRoom'].isnull()]
df['bedRoom'].head(2)
```

```python
## Numerical Value filter in bedRoom Column
df['bedRoom'] = df['bedRoom'].str.split(' ').str.get(0).astype('int')
df.head()
```

```python
# bathroom
df['bathroom'].value_counts()
```

```python
df['bathroom'].isnull().sum()
```

```python
df['bathroom'].head(2)
```

```python
## Numerical Value filter in bathroom Column
df['bathroom'] = df['bathroom'].str.split(' ').str.get(0).astype('int')
df['bathroom'].head(2)
```

```python
# balcony
df['balcony'].value_counts()
```

```python
df['balcony'].isnull().sum()
```

```python
df['balcony'].head(2)
```

```python
## Numerical Value filter in balcony Column
df['balcony'] = df['balcony'].str.split(' ').str.get(0).str.replace('No','0')
df.head()
```

```python
# additionalRoom
df['additionalRoom'].value_counts()
```

```python
## Filling Null Value with 'not available'
df['additionalRoom'].fillna('not available',inplace=True)
## Lower case 
df['additionalRoom'] = df['additionalRoom'].str.lower()
df['additionalRoom'].head(5)
```

```python
# floors
df['noOfFloor'].value_counts()
```

```python
df['noOfFloor'].isnull().sum()
```

```python
df['noOfFloor'].head(2)
```

```python
## Numerical Value filter in noOfFloor Column
df['noOfFloor'] = df['noOfFloor'].str.split(' ').str.get(0)
df['noOfFloor'].head(2)
```

```python
df.rename(columns={'noOfFloor':'floorNum'},inplace=True)
```

```python
## Filling NA for Null value
df['facing'].fillna('NA',inplace=True)
```

```python
## Add Both Columns area and property_Type
df['area'] = round((df['price']*10000000)/df['price_per_sqft'])
df.insert(loc=1,column='property_type',value='house')
```
